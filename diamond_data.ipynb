```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set() # Apply the default seaborn theme, scaling, and color palette

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
diamonds = sns.load_dataset("diamonds")  # Loaded in a Pandas DataFrame
diamonds.head()
```

```python
diamonds.describe()
```

```python
# List of categorical columns
categorical_cols = diamonds.select_dtypes(include='category').columns

# Display frequency table for each
for col in categorical_cols:
    print(f"\nFrequency table for '{col}':")
    print(diamonds[col].value_counts())
```

```python
# Plot for 'cut' column
sns.countplot(data=diamonds, x='cut')
plt.title('Count Plot for Cut')
plt.show()
```

```python
diamonds.isna().sum()
```

```python
# Scatter plot for two continuous variables: 'carat' vs 'price'
sns.scatterplot(data=diamonds, x='carat', y='price')
plt.title('Scatter Plot of Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```
